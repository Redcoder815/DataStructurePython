```
class Queue:
  def __init__(self, k):
    self.k = k
    self.queue = [None] * k
    self.head = self.tail = -1

  def isEmpty(self):
    return self.head == -1 and self.tail == -1

  def isFull(self):
    return self.tail == self.k -1

  def enqueue(self, item):
    if self.isFull():
      print('Queue is full')
    elif self.isEmpty():
      self.head = 0
      self.tail = 0
      self.queue[self.head] = item
    else:
      self.queue[self.tail+1] = item
      self.tail += 1

  def dequeue(self):
    if self.isEmpty():
      print('Queue is empty')
    elif self.tail == self.head:
      tem = self.queue[self.head]
      self.tail = -1
      self.head = -1
      return tem
    else:
      tem = self.queue[self.head]
      self.head += 1
      return tem

  def peek(self):
    if self.isEmpty():
      return 'The queue is empty'
    else:
      return self.queue[self.head]

  def printQueue(self):
    if self.isEmpty():
      print('The queue is empty')
    else:
      for i in range(self.head, self.tail + 1):
        print(f'{self.queue[i]} ')

model = Queue(5)
model.enqueue(1)
model.enqueue(2)
model.enqueue(3)
model.enqueue(4)
model.enqueue(5)
model.printQueue()
model.dequeue()
print('After dequeue')
model.printQueue()
```
