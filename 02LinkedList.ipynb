```
class Node:
  def __init__(self, item):
    self.item = item
    self.next = None

class LinkedList:
  def __init__(self):
    self.head = None

  def isEmpty(self):
    return self.head == None

  def insertion(self, value):
    if self.isEmpty():
      self.head = Node(value)
    else:
      temp = self.head
      while temp.next != None:
        temp = temp.next
      temp.next = Node(value)

  def deleteNode(self, position):
    if self.head is None:
      return
    current = self.head

    if position == 0:
      self.head = current.next
    currentPosition = 0
    while currentPosition < position - 1:
      current = current.next
      currentPosition += 1
      if current is None:
        break

    if current is None or current.next is None:
      return

    nodeToDelete = current.next
    current.next = nodeToDelete.next
    return self.head

  def search(self, value):
    if self.isEmpty():
      return 'List is empty'
    current = self.head
    while current is not None:
      if current.item == value:
        return 'Value found'
      current = current.next
    return 'Value is not found'

  def sortLinkedList(self):
    if self.isEmpty():
      return
    current = self.head
    indexNode = Node(None)

    while current is not None:
      indexNode = current.next
      while indexNode is not None:
        if current.item > indexNode.item:
          current.item,  indexNode.item = indexNode.item, current.item
        indexNode = indexNode.next
      current = current.next

  def printLinkedList(self):
    current = self.head
    while current:
      print(current.item, end=' ')
      current = current.next
```

```
my_list = LinkedList()
print('Is list empty?', my_list.isEmpty())

my_list.insertion(10)
my_list.insertion(20)
my_list.insertion(30)
my_list.insertion(40)

print('Is list empty?', my_list.isEmpty())
print('Linked List elements:')
my_list.printLinkedList()

my_list.insertion(50)
print('Linked List after adding 50:')
my_list.printLinkedList()
my_list.deleteNode(0)
my_list.deleteNode(2)
my_list.deleteNode(5)
my_list.deleteNode(2)
print('After delete')
my_list.printLinkedList()
my_list.search(20)
my_list.search(70)
my_list.insertion(0)
my_list.insertion(-75)
my_list.insertion(2)
my_list.insertion(1)
my_list.printLinkedList()
my_list.sortLinkedList()
print('After sort')
my_list.printLinkedList()
```
