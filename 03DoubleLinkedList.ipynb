```
class Node:
  def __init__(self, item):
    self.item = item
    self.prev = None
    self.next = None

class DoubleLinkedList:
  def __init__(self):
    self.head = None

  def isEmpty(self):
    return self.head is None

  def insertion(self, value):
    newNode = Node(value)
    if self.isEmpty():
      self.head = newNode
    else:
      current = self.head
      while current.next is not None:
        current = current.next
      current.next = newNode
      newNode.prev = current
    return self.head

  def insertBegin(self, value):
    newNode = Node(value)
    if self.isEmpty():
      self.head = newNode
    else:
      self.head.prev = newNode
      newNode.next = self.head
      self.head = newNode # Update head to the new node
    return self.head

  def insertAtPosition(self, position, value):
    newNode = Node(value)

    if position < 0:
      return self.head

    if self.isEmpty():
      self.head = newNode
      return self.head

    if position == 0:
      self.head.prev = newNode
      newNode.next = self.head
      self.head = newNode
    else:
      current = self.head
      currentPosition = 0
      while current.next is not None and currentPosition < position - 1:
        current = current.next
        currentPosition += 1

      if current.next is None:
        current.next = newNode
        newNode.prev = current
      else:
        newNode.next = current.next
        newNode.prev = current
        current.next.prev = newNode
        current.next = newNode
    return self.head

  def deleteNode(self, position):
    if self.isEmpty():
      print('List is empty')
      return self.head

    if position == 0:
      if self.head.next is None:
        self.head = None
      else:
        self.head = self.head.next
        self.head.prev = None
      return self.head

    current = self.head
    currentPosition = 0
    while current.next is not None and currentPosition < position - 1:
      current = current.next
      currentPosition += 1

    if current.next is None:
      print(f"Error: Position {position} is out of bounds for deletion. Node not found.")
      return self.head
    else:
      nodeToDelete = current.next
      next_node = nodeToDelete.next

      current.next = next_node

      if next_node is not None:
        next_node.prev = current
      nodeToDelete.prev = None
      nodeToDelete.next = None
      return self.head

  def printList(self):
    if self.isEmpty():
      print('The list is empty')
    else:
      current = self.head
      while current is not None:
        print(current.item)
        current = current.next

dLinked = DoubleLinkedList()
dLinked.insertion(1)
dLinked.insertion(0)
dLinked.insertion(9)
dLinked.insertion(6)
dLinked.insertion(4)
# dLinked.insertAtPosition(0, 2)
# dLinked.insertAtPosition(2, 30)
dLinked.printList()
dLinked.deleteNode(4)
print("After delelete")
dLinked.printList()
```
