```python
my_list = [None]*10
```

```python
print(my_list)
```

```python
def hash_function(name):
  name_to_unicode = 0
  for c  in name:
    name_to_unicode += ord(c)
  index = name_to_unicode % 10
  return index

def add_name(name):
  index = hash_function(name)
  my_list[index] = name

def is_name_exist(name):
  index = hash_function(name)
  return my_list[index] == name
```

```python
add_name('Bob')
add_name('Pete')
add_name('Jones')
add_name('Lisa')
add_name('Siri')
```

```python
print(my_list)
```

```python
print('Pete name is exits', is_name_exist('Pete'))
```

```python
another_list = [[] for _ in range(10)]
```

```python
print(another_list)
```

```python
def hash_function(name):
  name_to_unicode = 0
  for c  in name:
    name_to_unicode += ord(c)
  index = name_to_unicode % 10
  return index

def add_name(name):
  index = hash_function(name)
  another_list[index].append(name)

def is_name_exist(name):
  index = hash_function(name)
  return name in another_list[index]
```

```python
add_name('Bob')
add_name('Pete')
add_name('Jones')
add_name('Lisa')
add_name('Siri')
add_name('Stuart')
```

```python
print(another_list)
```

```python
print('Is Pete in another_list?', is_name_exist('Pete'))
print('Is Bob in another_list?', is_name_exist('Bob'))
print('Is John in another_list?', is_name_exist('John'))
print('Is Stuart in another_list?', is_name_exist('Stuart'))
```
