```
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def iterative_inorder(root):
    stack = []
    current = root

    while stack or current:
        while current:
            stack.append(current)
            current = current.left

        current = stack.pop()
        print(current.value, end=" ")

        current = current.right

def inorder(root):
  if root:
    inorder(root.left)
    print(root.value, end = ' ')
    inorder(root.right)

def iterative_postorder(root):
  if not root:
    return

  s1 = []
  s2 = []

  s1.append(root)

  while s1:
    node = s1.pop()
    s2.append(node)

    if node.left:
      s1.append(node.left)
    if node.right:
      s1.append(node.right)

  while s2:
    node = s2.pop()
    print(node.value, end=' ')

def postorder(root):
  if root:
    postorder(root.left)
    postorder(root.right)
    print(root.value, end = ' ')

def preorder(root):
   if root:
      print(root.value, end = ' ')
      preorder(root.left)
      preorder(root.right)

def iterative_preorder(root):
  if not root:
    return
  stack = []
  stack.append(root)
  while stack:
    node = stack.pop()
    print(node.value, end = ' ')

    if node.right:
      stack.append(node.right)
    if node.left:
      stack.append(node.left)

def iterativeIsFullBinaryTree(root):
  if root is None:
    return True

  queue = [root]

  while queue:
    node = queue.pop(0)

    if (node.left is None and node.right is not None) or (node.left is not None and node.right is None):
      return False

    if node.left:
      queue.append(node.left)
    if node.right:
      queue.append(node.right)

  return True

def isFullTree(root):
  if root is None:
    return True
  if root.left is None and root.right is None:
    return True
  if root.left is not None and root.right is not None:
    return (isFullTree(root.left) and isFullTree(root.right))
  return False

# Example Usage:
#      1
#     / \
#    2   3
#   / \
#  4   5

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

iterative_inorder(root)
print('\n')
inorder(root)
print('\n')
postorder(root)
print('\n')
iterative_postorder(root)
print('\n')
preorder(root)
print('\n')
iterative_preorder(root)
print('\n')
print(iterativeIsFullBinaryTree(root))
print(isFullTree(root))
```
