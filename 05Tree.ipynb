```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def iterative_inorder(root):
    stack = []
    current = root

    while stack or current:
        while current:
            stack.append(current)
            current = current.left

        current = stack.pop()
        print(current.value, end=" ")

        current = current.right

def inorder(root):
  if root:
    inorder(root.left)
    print(root.value, end = ' ')
    inorder(root.right)

def iterative_postorder(root):
  stack = []
  current = root

  while stack or current:
    while current:
      stack.append(current)
      current = current.right
    current = stack.pop()
    print(current.value, end = ' ')
    current = current.left

def postorder(root):
  if root:
    postorder(root.left)
    postorder(root.right)
    print(root.value, end = ' ')


# Example Usage:
#      1
#     / \
#    2   3
#   / \
#  4   5

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

iterative_inorder(root)
print('\n')
inorder(root)
print('\n')
postorder(root)
print('\n')
iterative_postorder(root)
```
