```
def BinarySearch(array, value):
  array = arraySorted(array)
  low = 0
  high = len(array)-1
  while low <= high:
    mid = int((low+high)/2)
    if array[mid] == value:
      return True
    elif array[mid] > value:
      high = mid - 1
    else:
      low = mid + 1
  return False
```

```
def arraySorted(array):
  l = len(array)
  for i in range(l):
    for j in range(i + 1, l):
      if array[j]<array[i]:
        array[j], array[i] = array[i], array[j]
  return array
```

```
# array = [3, 4, 5, 6, 7, 8, 9]
array = [4, 1, 3, 6, 0, -1]
result = BinarySearch(array, -1)
if result is True:
  print('Value has found')
else:
  print('Value not found')
```
