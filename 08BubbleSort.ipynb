BubbleSort

```
def bubbleSort(array):
  l = len(array)
  for i in range(l):
    swap = False
    for j in range(i + 1, l):
      if array[j]<array[i]:
        array[j], array[i] = array[i], array[j]
        swap = True
    if swap is False:
      break
  return array
```

```
arr = [9,-9,0,2,1,5,2]
print(bubbleSort(arr))
```
