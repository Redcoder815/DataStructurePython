```
def selectionSort(array):
  length = len(array)
  for i in range(length):
    min_value = array[i]
    min_value_index = i
    for j in range(i+1, length):
      if min_value > array[j]:
        min_value = array[j]
        min_value_index = j
    array[i], array[min_value_index] = array[min_value_index], array[i]
  return array
```

```
data = [-2, 45, 0, 11, -9]
print(selectionSort(data))
```
