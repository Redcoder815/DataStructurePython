```
def insertionSort(array):
  length = len(array)
  for i in range(1, length):
    key = array[i]
    j = i - 1
    while j>=0 and key < array[j]:
      array[j + 1] = array[j]
      j = j - 1
    array[j+1] = key
  return array
```

```
data = [-2, 45, 0, 11, -9]
print(insertionSort(data))
```
