```
def merge(array, left_size, mid, right_end):
    n = mid - left_size + 1
    m = right_end - mid

    left_array = [0] * n
    right_array = [0] * m

    for i in range(n):
        left_array[i] = array[i+left_size]
    for j in range(m):
        right_array[j] = array[mid+1+j]

    i = 0
    j = 0
    k = left_size

    while i < n and j < m:
        if left_array[i] <= right_array[j]:
            array[k] = left_array[i]
            i = i+1
        else:
            array[k] = right_array[j]
            j = j+1
        k = k + 1

    while i < n:
        array[k] = left_array[i]
        i= i+1
        k = k + 1

    while j < m:
        array[k] = right_array[j]
        j = j + 1
        k = k + 1



def merge_sort(array):
    n = len(array)
    current_size = 1

    while current_size < n:
        left_start = 0

        while left_start < n - 1:
            mid = min(left_start + current_size - 1, n - 1)
            right_end = min(left_start + 2 * current_size - 1, n - 1)

            merge(array, left_start, mid, right_end)
            left_start += 2 * current_size
        current_size *= 2

array = [6, 5, 12, 10, 9, 1]
merge_sort(array)
print(array)
```
