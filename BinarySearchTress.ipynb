```python
class TreeNode:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

def search(node, target):
  if node is None:
    return None
  elif node.data == target:
    return node
  elif node.data < target:
    return search(node.left, target)
  else:
    return search(node.right, target)

def insert(node, value):
  if node is None:
    return TreeNode(value)
  else:
    if value < node.data:
      node.left = insert(node.left, value)
    else:
      node.right = insert(node.right, value)
    return node

def findLowest(node):
  minValue = node.data
  while node.left:
    node = node.left
    minValue = node.data
  return minValue

def deleteNode(root, nodeToDelete):
  current = root
  previous = None
  while current and current.data != nodeToDelete:
    previous = current
    if nodeToDelete < current.data:
      current = current.left
    else:
      current = current.right

  if current is None:
    return root
  # if root has one child in left or right
  if current.left is None or current.right is None:
    new_current = current.left if current.left else current.right

    if previous is None:
      return new_current

    if current == previous.left:
      previous.left = new_current
    else:
      previous.right = new_current
#       # Case 3: Node has two children
  else:
    previous_successor = current
    successor = current.right
    while successor.left:
      previous_successor = successor
      successor = successor.left

    current.data = successor.data

    if previous_successor != current:
      previous_successor.left = successor.right
    else:
      previous_successor.right = successor.right
  return root

# root = TreeNode(13)
# node7 = TreeNode(7)
# node15 = TreeNode(15)
# node3 = TreeNode(3)
# node8 = TreeNode(8)
# node14 = TreeNode(14)
# node19 = TreeNode(19)
# node18 = TreeNode(18)

# root.left = node7
# root.right = node15

# node7.left = node3
# node7.right = node8

# node15.left = node14
# node15.right = node19

# node19.left = node18

# nodeValue = search(root, 13)

# print(f'value {nodeValue.data}')

# insert(root, 10)
# insert(root, 1)
# insert(root, 0)

root = TreeNode(50)
root.left = TreeNode(30)
root.right = TreeNode(70)
root.right.left = TreeNode(60)
root.right.right = TreeNode(80)
root.right.left.right = TreeNode(65)
deleteNode(root,80)
print(findLowest(root))
```
