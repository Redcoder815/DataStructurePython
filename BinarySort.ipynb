```python
array = [1, 3, 0, 7, 8, 7, 3, 11, 10, 14, 17, 15]

for i in range(0, len(array) - 1):
  for j in range( i+1, len(array)):
    if array[i] > array[j]:
      array[i], array[j] = array[j], array[i]

print(array)

def binary_sort(array, value):
  left = 0
  right = len(array) - 1
  while left <= right:
    mid = (left + right) // 2

    if array[mid] == value:
      return mid

    if array[mid] < value:
      left = mid + 1
    else:
      right = mid - 1

  return -1

result = binary_sort(array, 3)
print(result)
```
