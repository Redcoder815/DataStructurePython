```python
mylist = [64, 34, 25, 3, 12, 22, 11, 90, 5]

n = len(mylist)

for i in range(n-1):
  swap = False
  for j in range(n-i-1):
    if mylist[j]>mylist[j+1]:
      mylist[j], mylist[j+1] = mylist[j+1], mylist[j]
      swap = True
  if not swap:
    break

print(mylist)
```
