```python
mylist = [ 2, 3, 0, 2, 3, 2]
# mylist = [4, 2, 2, 6, 3, 3, 1, 6, 5, 2, 3]


def counting_sort(array):
    maximum_value = max(array)
    counting_index_array = [0] * (maximum_value + 1)
    for i in range(len(array)):
        for j in range(len(array)):
            if i == array[j]:
                counting_index_array[i] = counting_index_array[i] + 1
    # print(counting_index_array)
    new_array=[]
    for idx, k in enumerate(counting_index_array):
        if k>0:
         for p in range(0,k):
             new_array.append(idx)

    return new_array

print(counting_sort(mylist))
```

```python
def countingSort(arr):
  max_val = max(arr)
  count = [0] * (max_val + 1)

  while len(arr) > 0:
    num = arr.pop(0)
    count[num] += 1

  for i in range(len(count)):
    while count[i] > 0:
      arr.append(i)
      count[i] -= 1

  return arr

mylist = [4, 2, 2, 6, 3, 3, 1, 6, 5, 2, 3]
mysortedlist = countingSort(mylist)
print(mysortedlist)
```
