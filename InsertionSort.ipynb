```python
mylist = [64, 34, 25, 12, 22, 11, 90, 5]

n = len(mylist)

for i in range(1,n):
  sorted_index = i
  minValue = mylist.pop(i)
  for j in range(i-1, -1, -1):
    if mylist[j]>minValue:
      sorted_index = j
  mylist.insert(sorted_index, minValue)

print(mylist)
```

```python
mylist = [64, 34, 25, 12, 22, 11, 90, 5]

n = len(mylist)

for i in range(1,n):
  insert_index = i
  minValue= mylist[i]

  for j in range(i-1, -1, -1):
    if mylist[j]>minValue:
      mylist[j+1] = mylist[j]
      insert_index = j
    else:
      break
  mylist[insert_index] = minValue

print(mylist)
```
