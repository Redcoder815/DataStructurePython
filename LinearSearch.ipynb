```python
my_list = [7, 8 ,9 ,10,2,4,0,1]

if 4 in my_list:
  print('Found')
else:
  print('Not found')

def linearSearch(my_list, value):
  for i in range(len(my_list)):
    if my_list[i] == value:
      return i
    return -1

result = linearSearch(my_list, 4)

if result != -1:
  print('Result found')
else:
  print('Not found')
```
