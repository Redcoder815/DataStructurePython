```python
mylist = [64, 34, 25, 5, 22, 11, 90, 12]

def pivot_index(array, low, high):
  pivot = array[high]
  index = low - 1

  for j in range(low, high):
    if array[j] <= pivot:
      index = index + 1
      array[index], array[j] = array[j], array[index]
  array[index + 1], array[high] = array[high], array[index+1]
  return index+1

def partition(array, low = 0, high = None):
  if high is None:
    high = len(array) - 1

  if low < high:
    pivot_idx = pivot_index(array, low, high)
    partition(array, low, pivot_idx-1)
    partition(array, pivot_idx + 1, high)

partition(mylist)
print(mylist)
```
