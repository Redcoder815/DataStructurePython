```python
# myArray = [ 33, 45, 40, 25, 17, 24]
# myArray = [170, 45, 75, 90, 802, 24, 2, 66]
myArray = [170, 45, 75]


def radix_sort(array):
    radixArray = [[], [], [], [], [], [], [], [], [], []]

    maxValue = max(array)
    exp = 1

    while maxValue // exp > 0:
        while len(array) > 0:
            value = array.pop()
            digit_index = (value // exp) % 10
            radixArray[digit_index].append(value)

        for nested_array in radixArray:
            while len(nested_array) > 0:
                sorted_value = nested_array.pop()
                array.append(sorted_value)

        exp *= 10
radix_sort(myArray)

print(myArray)
```
