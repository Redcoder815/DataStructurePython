```python
my_list = [ 7, 12, 9, 11, 3]

for i in range(len(my_list)-1):
  min_index = i
  for j in range(i+1, len(my_list)):
    if my_list[j] < my_list[min_index]:
      min_index = j
  my_list[min_index], my_list[i] = my_list[i], my_list[min_index]

print(my_list)
```

```python
my_list = [ 7, 12, 9, 11, 3]
for i in range(len(my_list)-1):
  min_index = i
  for j in range(i+1, len(my_list)):
    if my_list[j] < my_list[min_index]:
      min_index = j
  min_value = my_list.pop(min_index)
  my_list.insert(i, min_value)

print(my_list)

```
