```python
class Stack:
  def __init__(self):
    self.stack = []

  def push(self, element):
    self.stack.append(element)

  def peek(self):
    if self.isEmpty():
      return "Stack is empty"
    return self.stack[-1]

  def pop(self):
    if self.isEmpty():
      return 'Stack is empty'
    return self.stack.pop()

  def isEmpty(self):
    return len(self.stack) == 0

  def size(self):
    return len(self.stack)

Stack1 = Stack()
Stack1.push(1)
Stack1.push(2)
Stack1.push(3)

print(Stack1.peek())
print(Stack1.pop())
print(Stack1.peek())
print(Stack1.isEmpty())
```
