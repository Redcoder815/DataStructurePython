```python
class Node:
  def __init__(self,value):
    self.value = value
    self.next = None

class Stack:
  def __init__(self):
    self.head = None
    self.size = 0

  def push(self, value):
    newNode = Node(value)
    if self.head:
      newNode.next = self.head
    self.head = newNode
    self.size += 1

  def peek(self):
    if self.isEmpty():
      return "stack is empty"
    return self.head.value

  def pop(self):
    if self.isEmpty():
      return "stack is empty"
    popNode = self.head
    self.head = self.head.next
    self.size -= 1
    return popNode.value

  def isEmpty(self):
    return self.size == 0

  def stackSize(self):
    return self.size

  def taverseAndPrint(self):
    currentNode = self.head
    while currentNode:
      print(currentNode.value)
      currentNode = currentNode.next

myStack = Stack()
myStack.push('A')
myStack.push('B')
myStack.push('C')

print("LinkedList: ", end="")
myStack.taverseAndPrint()
print("Peek: ", myStack.peek())
print("Pop: ", myStack.pop())
print("LinkedList after Pop: ", end="")
myStack.taverseAndPrint()
print("isEmpty: ", myStack.isEmpty())
print("Size: ", myStack.stackSize())
```
